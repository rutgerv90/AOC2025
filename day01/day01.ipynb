```python
import numpy as np
import re
import json
import math

# Load input
with open("sample.txt") as f:
    sample_data = []
    data = f.read().splitlines()
    for line in data:
        letter = line[0]
        nr = int(line[1:])
        sample_data.append([letter, nr])

# Load input data
with open("input.txt") as f:
    input_data = []
    data = f.read().splitlines()
    for line in data:
        letter = line[0]
        nr = int(line[1:])
        input_data.append([letter, nr])
```

## ‚≠ê Part 1

```python
# Part 1 solution


# start at 50; range = 0 -> 99
def part1(data):
    target = 50
    password = 0
    for rotation in data:
        if rotation[0] == "L":
            target -= rotation[1]
        else:
            target += rotation[1]

        # after rotation modulo by 100
        target = target % 100

        if target == 0:
            password += 1
    return password


print(part1(input_data))
```

## ‚≠ê Part 2

```python
# Part 2 solution
def part2(data):
    target = 50
    password = 0
    for rotation in data:
        for i in range(0, rotation[1]):
            if rotation[0] == "L":
                target -= 1
            else:
                target += 1
            if target % 100 == 0:
                password += 1

        # after rotation modulo by 100
        target = target % 100
    return password


print(part2(input_data))
```
