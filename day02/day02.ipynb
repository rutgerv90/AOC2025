```python
import numpy as np
import re
import json
import math

# Load sample data
with open("sample.txt") as f:
    data = f.read().split(",")
    sample_data = []
    for ids in data:
        a, b = ids.split("-")
        sample_data.append([int(a), int(b)])


# Load input data
with open("input.txt") as f:
    data = f.read().split(",")
    input_data = []
    for ids in data:
        a, b = ids.split("-")
        input_data.append([int(a), int(b)])
```

## ‚≠ê Part 1

```python
# Part 1 solution
def part1(data):
    score = 0
    for line in data:
        # create array of the range
        nums = np.arange(line[0], line[1] + 1)
        # invalid ids can only occur in even length numbers; filter out uneven
        filtered = np.array([x for x in nums if remove_uneven(x)])

        # loop over filtered nums
        for num in filtered:
            # cast to string
            z = str(num)
            # find the midpoint
            mid = len(z) // 2
            # divide into 2 parts
            part1, part2 = z[:mid], z[mid:]
            # up the score if parts match
            if part1 == part2:
                score += num
    return score


def remove_uneven(x):
    return len(str(x)) % 2 == 0


print("score = ", part1(input_data))
```

## ‚≠ê Part 2

```python
# Part 2 solution
def part2(data):
    score = 0
    for line in data:
        # create array of the range
        nums = np.arange(line[0], line[1] + 1)

        # loop over numbers
        for item in nums:
            y = str(item)
            z = len(str(item))
            # always check if item consists of repeating 1char
            # EXCLUDE single digits
            if len(set(y)) == 1 and z > 1:
                score += item
            else:
                # loop over divisors
                for j in divisors(z):
                    # create a temp aray
                    temp = []
                    # in range of divisors split string into chunks
                    for i in range(0, z, j):
                        temp.append(y[i : i + j])
                    # if the set repeats add to score
                    if len(set(temp)) == 1:
                        score += item
                        # make sure its not counted twice
                        break

    return score


# return divisors BETWEEN 1 and 'n'
def divisors(n):
    return [i for i in range(2, n) if n % i == 0]


print("score =", part2(input_data))
```
