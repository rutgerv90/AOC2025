```python
import numpy as np
import re
import json
import math


# Load sample
with open("sample.txt") as f:
    data = f.read().splitlines()
    data = [list(map(int, list(line))) for line in data]
    sample_data = np.array(data)

# Load input
with open("input.txt") as f:
    data = f.read().splitlines()
    data = [list(map(int, list(line))) for line in data]
    input_data = np.array(data)
```

## ‚≠ê Part 1

```python
# Part 1 solution
def part1(data):
    score = 0
    for line in data:
        # sorted_line = np.sort(line)[::-1]
        # find the highest number on the left side
        left = line[:-1]
        right = line[1:]
        left_max = np.max(left)
        # get the index of the highest leftmost number
        max_index = np.where(left == left_max)[0]
        first_num = line[max_index[0]]
        # the second num is the highest number right of the leftmost
        second_num = np.max(line[max_index[0] + 1 :])
        # combine the numbers
        line_score = int(str(first_num) + str(second_num))
        # increase the score
        score += line_score
    return score


print("score =", part1(input_data))
```

## ‚≠ê Part 2

```python
# Part 2 solution
def part2(data):
    # init score
    score = 0
    # go through lines in data
    for line in data:
        # out string (will be converted to int later)
        out = ""
        # go through 12 chars
        for i in range(12):
            # range based on position in the list
            x = line[: len(line) - (11 - i)]
            # find max number in sliced list
            num_max = max(x)
            # get the index of the max number (first one is leftmost)
            pos_max = np.where(line == num_max)[0][0]
            # append max num to the out str
            out += str(num_max)
            # slice the remaining line based on index of found max
            line = line[pos_max + 1 :]
        # update the score
        score += int(out)
    return score


print("score =", part2(input_data))
```
