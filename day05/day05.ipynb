```python
import numpy as np
import re
import json
import math
import copy

# Load sample
with open("sample.txt") as f:
    data = f.read().splitlines()
    sample_range = []
    sample_data = []
    for line in data:
        if "-" in line:
            a, b = line.split("-")
            sample_range.append([int(a), int(b)])
        elif not line:
            continue
        else:
            sample_data.append(int(line))

# Load input
with open("input.txt") as f:
    data = f.read().splitlines()
    input_range = []
    input_data = []
    for line in data:
        if "-" in line:
            a, b = line.split("-")
            input_range.append([int(a), int(b)])
        elif not line:
            continue
        else:
            input_data.append(int(line))
```

## â­ Part 1

```python
# Part 1 solution
def part1(ingredients, xrange):
    # init score
    fresh = 0
    # go over ids
    for id in ingredients:
        for r in xrange:
            # check if theyre in range and if so , up the score, break and move on
            if id >= r[0] and id <= r[1]:
                fresh += 1
                break
    # return score
    return fresh


print("Part 1 score =", part1(input_data, input_range))
```

## â­ Part 2

```python
# Part 2 solution
def part2(xrange):
    # init score
    fresh = 0
    # copy range to avoid altering original list
    intervals = copy.deepcopy(xrange)
    # sort the range (defaults to sorting by min nr which is what we want)
    intervals.sort()
    # init a combined list
    combined = []
    # go over items in the intervals
    for item in intervals:
        # compare them to the previous checked items
        # if no previous, add it
        # OR check if the last interval ends before the current one starts (watch the gap)
        if not combined or combined[-1][1] < item[0] - 1:
            # append to combined list
            combined.append(item)
        else:
            # if they DO overlap, combine them into 1 by setting the new max
            combined[-1][1] = max(combined[-1][1], item[1])

    # count the ranges
    for item in combined:
        # max - min (+1 for range inclusivity)
        fresh += item[1] - item[0] + 1
    return fresh


print("Part 2 score =", part2(input_range))
```
