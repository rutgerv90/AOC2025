```python
import numpy as np
import re
import json
import math

# Load sample
with open("sample.txt") as f:
    data = f.read().splitlines()
    sample_data = []
    sample_data_string = data[:-1]
    for line in data[:-1]:
        sample_data.append([int(item) for item in line.split(" ") if item != ""])
    for line in data[-1:]:
        sample_operations = [item for item in line.split(" ") if item != ""]

# Load input
with open("input.txt") as f:
    data = f.read().splitlines()
    input_data = []
    input_data_string = data[:-1]
    for line in data[:-1]:
        input_data.append([int(item) for item in line.split(" ") if item != ""])
    for line in data[-1:]:
        input_operations = [item for item in line.split(" ") if item != ""]
```

## ‚≠ê Part 1

```python
# Part 1 solution
def part1(data, ops):
    score = 0
    # transpose data so cols are now rows
    data = np.transpose(data)
    # loop over them
    for idx, row in enumerate(data):
        # match the right ops
        match ops[idx]:
            case "+":
                # sum the row
                score += np.sum(row)
            case "*":
                # get product of row
                score += np.prod(row)
            case _:
                # there should be no other ops in part 1 at least
                break
    # return score
    return score


print("Part 1 score:", part1(input_data, input_operations))
```

## ‚≠ê Part 2

```python
# Part 2 solution
def part2(data, ops):
    # init score
    score = 0
    # holds data with numbers
    hold = []
    # go through the string backwards, char by char
    for i in range(len(data[0]) - 1, -1, -1):
        # init empty string
        num = ""
        for row in data:
            if row[i] != " ":
                # append digit to string if found
                num += row[i]
        # append to string, '' stays to seperate blocks
        hold.append(num)
    # reverse list of operations
    rev_ops = ops[::-1]
    # add extra '' at the end, used later
    hold.append("")
    # init temp array
    temp = []
    # var to keep track of index
    a = 0
    for idx, item in enumerate(hold):
        # if a seperator is found, score the previous block
        if item == "":
            print(rev_ops[a], temp)
            # same match as in part 1 but reversed
            match rev_ops[a]:
                case "+":
                    # sum the row
                    score += np.sum(temp)
                case "*":
                    # get product of row
                    score += np.prod(temp)
                case _:
                    # there should be no other ops in part 1 at least
                    break
            # up the index
            a += 1
            # clear the temp array
            temp = []
        else:
            # add the integer to the temp array
            temp.append(int(item))

    # return score
    return score


# print(sample_data_string)
print("Part 2 score:", part2(input_data_string, input_operations))
```
