```python
import numpy as np
import re
import json
import math
import copy
import random

# Load sample
with open("sample.txt") as f:
    data = f.read().splitlines()
    sample_data = np.array([list(line) for line in data])

# Load input
with open("input.txt") as f:
    data = f.read().splitlines()
    input_data = np.array([list(line) for line in data])
```

## ‚≠ê Part 1

```python
# Part 1 solution
def part1(data):
    rows = len(data)
    cols = len(data[0])
    out = data.copy()
    splits = 0
    # skip first row
    for row in range(1, rows):
        for col in range(0, cols):
            if out[row][col] == ".":
                if out[row - 1][col] == "|" or out[row - 1][col] == "S":
                    out[row][col] = "|"
            elif out[row][col] == "^":
                if out[row - 1][col] == "|":
                    splits += 1
                    out[row][col - 1] = "|"
                    out[row][col + 1] = "|"
    return splits


print("Part 1 score =", part1(input_data))
```
