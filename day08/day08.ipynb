```python
import numpy as np
import re
import json
import math
import copy
from collections import Counter

# Load sample
with open("sample.txt") as f:
    sample_data = [line.split(",") for line in f.read().splitlines()]

# Load input
with open("input.txt") as f:
    input_data = [line.split(",") for line in f.read().splitlines()]
```

## â­ Part 1

```python
# Part 1 solution
def part1(data, connections):
    # make copy to not change input lsit
    data = copy.deepcopy(data)
    # keeps track of later connections between boxes
    tracker = np.zeros(len(data))
    # json that will hold the distances between all junctions box combinations
    distances = {}
    # init box nr
    box_nr = 0
    while len(data) > 1:
        # pop out the box from the list
        box = list(map(int, data.pop(0)))
        # loop over the rest
        for idx, j in enumerate(data):
            # number the box based on offset
            other_box_nr = box_nr + idx + 1
            # convert to int
            other_box = list(map(int, j))
            # use np to calc the distance
            box_distance = np.linalg.norm(np.array(box) - np.array(other_box))
            # add the distance as a key, and the box nr as the values
            distances[float(box_distance)] = {"from": box_nr, "to": other_box_nr}
        box_nr += 1
    # sort the distances json
    out = dict(sorted(distances.items()))
    # init a number
    z = 1
    # go over the first x connections (sample = 10, input = 1000)
    for key, value in list(out.items())[:connections]:
        # print(key, "->", value)
        # keep track of to/from box nr
        _to = value["to"]
        _from = value["from"]
        # if both are already numbered, all connections are now part of a single loop
        if tracker[_to] > 0 and tracker[_from] > 0:
            tracker[tracker == tracker[_to]] = tracker[_from]
        # if the FROM box is unassigned, assign it to the TO nr
        elif tracker[_to] > 0 and tracker[_from] == 0:
            tracker[_from] = tracker[_to]
        # if the TO box is unassigned, assign it to the FROM nr
        elif tracker[_from] > 0 and tracker[_to] == 0:
            tracker[_to] = tracker[_from]
        # if both boxes do not already exists in a group make a new group
        else:
            tracker[_to] = z
            tracker[_from] = z
        # up the tracking nr
        z += 1
    # make a counter of the connection tracker
    counts = Counter(tracker)
    # remove 0 (unassigned)
    del counts[0]
    # init temp array
    temp = []
    # print(json.dumps(counts, indent=2))
    # loop over the 3 most common/biggest groups
    for i in counts.most_common(3):
        # append the size
        temp.append(i[1])
    # return the product
    return np.prod(temp)


# print(part1(sample_data, 10))
print("Part 1 score =", part1(input_data, 1000))
```

```python
# Part 2 solution
def part2(data, connections):
    # make copy to not change input lsit
    data = copy.deepcopy(data)
    data2 = copy.deepcopy(data)
    # keeps track of later connections between boxes
    tracker = np.zeros(len(data))
    # json that will hold the distances between all junctions box combinations
    distances = {}
    # init box nr
    box_nr = 0
    while len(data) > 1:
        # pop out the box from the list
        box = list(map(int, data.pop(0)))
        # loop over the rest
        for idx, j in enumerate(data):
            # number the box based on offset
            other_box_nr = box_nr + idx + 1
            # convert to int
            other_box = list(map(int, j))
            # use np to calc the distance
            box_distance = np.linalg.norm(np.array(box) - np.array(other_box))
            # add the distance as a key, and the box nr as the values
            distances[float(box_distance)] = {"from": box_nr, "to": other_box_nr}
        box_nr += 1
    # sort the distances json
    out = dict(sorted(distances.items()))
    # init a number
    z = 1
    final_x_coords = []
    # go over the first x connections (sample = 10, input = 1000)
    for key, value in list(out.items()):
        # print(key, "->", value)
        # keep track of to/from box nr
        _to = value["to"]
        _from = value["from"]
        # if both are already numbered, all connections are now part of a single loop
        if tracker[_to] > 0 and tracker[_from] > 0:
            tracker[tracker == tracker[_to]] = tracker[_from]
        # if the FROM box is unassigned, assign it to the TO nr
        elif tracker[_to] > 0 and tracker[_from] == 0:
            tracker[_from] = tracker[_to]
        # if the TO box is unassigned, assign it to the FROM nr
        elif tracker[_from] > 0 and tracker[_to] == 0:
            tracker[_to] = tracker[_from]
        # if both boxes do not already exists in a group make a new group
        else:
            tracker[_to] = z
            tracker[_from] = z
        # IF the whole grid is one big set (part 2 criteria)
        if len(set(tracker)) == 1:
            print(
                "Found final connection between boxes",
                "\nBox:",
                _to,
                data2[_to],
                "and \nBox:",
                _from,
                data2[_from],
            )
            # look at the final boxes x coords
            final_x_coords.extend([int(data2[_from][0]), int(data2[_to][0])])
            # break the loop
            break
        # up the tracking nr
        z += 1
    # return the product of the x coords
    return np.prod(final_x_coords)


# print("Part 2 score =", part2(sample_data, 10))
print("Part 2 score =", part2(input_data, 1000))
```
