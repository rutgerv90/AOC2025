```python
import numpy as np
import re
import json
import math
import copy
from collections import Counter

# Load sample
with open("sample.txt") as f:
    sample_data = [list(map(int, line.split(","))) for line in f.read().splitlines()]

# Load input
with open("input.txt") as f:
    input_data = [list(map(int, line.split(","))) for line in f.read().splitlines()]
```

## â­ Part 1

```python
# Part 1 solution
def part1(data):
    # init max area
    max_area = 0
    # copy data
    data = copy.deepcopy(data)
    # loop over data
    while len(data) > 1:
        # remove first item from data
        tile = data.pop(0)
        # compare them against the others
        for idx, item in enumerate(data):
            # calculate area between them
            area = calculate_area(tile, item)
            # if the area is bigger than current max, set it as new
            if area > max_area:
                max_area = area
    # retun max area
    return max_area


# Calculate area
def calculate_area(tile, item):
    return (abs(tile[0] - item[0]) + 1) * (abs(tile[1] - item[1]) + 1)


print("Part 1 score =", part1(input_data))
```

```python
# Part 2 solution
def part2(data):
    # init max area
    max_area = 0
    # copy data
    data = copy.deepcopy(data)
    data2 = np.array(copy.deepcopy(data))
    grid_2d = np.zeros((max(data2[:, 1]) + 2, max(data2[:, 0]) + 2))
    prev_loc = []
    # print(data)
    # add the first entry to the la
    data2 = np.append(data2.copy(), [data2[0]], axis=0)
    # print(data)
    # grid_2d[1, 4:2] = 5
    for item in data2:
        if len(prev_loc) > 0:
            if item[1] == prev_loc[1]:
                grid_2d[
                    item[1],
                    min(item[0], prev_loc[0]) + 1 : max(item[0], prev_loc[0]) + 1,
                ] = 2
            else:
                grid_2d[
                    min(item[1], prev_loc[1]) + 1 : max(item[1], prev_loc[1]) + 1,
                    item[0],
                ] = 2
            grid_2d[prev_loc[1], prev_loc[0]] = 1
        grid_2d[item[1], item[0]] = 1
        prev_loc = item

    print(np.shape(grid_2d))
    filled_grid = flood_fill(grid_2d.copy(), (0, 0), target=0, replacement=9)
    # print(filled_grid)
    # raise
    # loop over data
    while len(data) > 1:
        # remove first item from data
        tile = data.pop(0)
        # compare them against the others
        for idx, item in enumerate(data):
            # calculate area between them
            area = calculate_area(tile, item, filled_grid)
            # if the area is bigger than current max, set it as new
            if area > max_area:
                max_area = area
    # retun max area
    return max_area


# Calculate area
def calculate_area(tile, item, filled_grid):
    # (part2) cast 4 rays
    # print(tile, item)
    # print(filled_grid[tile[1] : item[1] + 1, tile[0] : item[0] + 1])

    mini_grid = filled_grid[
        min(tile[1], item[1]) : max(tile[1], item[1]) + 1,
        min(tile[0], item[0]) : max(tile[0], item[0]) + 1,
    ]
    # print(mini_grid)

    if np.count_nonzero(mini_grid == 9):
        return 0

    # print("---")
    return (abs(tile[0] - item[0]) + 1) * (abs(tile[1] - item[1]) + 1)


# flood fill function taken from copilot
def flood_fill(arr, start, target, replacement):
    rows, cols = arr.shape
    r0, c0 = start

    if arr[r0, c0] != target or target == replacement:
        return arr

    stack = [(r0, c0)]
    while stack:
        r, c = stack.pop()
        if arr[r, c] == target:
            arr[r, c] = replacement
            # 4-neighbors (up, down, left, right)
            for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nr, nc = r + dr, c + dc
                if 0 <= nr < rows and 0 <= nc < cols:
                    stack.append((nr, nc))
    return arr


# works on sample data, input nrs too large
print("Part 2 score =", part2(input_data))
```
